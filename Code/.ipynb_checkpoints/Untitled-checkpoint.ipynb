```python
#Calendar Module
# Enter your code here. Read input from STDIN. Print output to STDOUT
import calendar
date = input().split()
m = int(date[0])
d = int(date[1])
y = int(date[2])
day = calendar.weekday(y, m, d)
print(list(calendar.day_name)[day].upper())

#Time Delta
#!/bin/python3

import math
import os
import random
import re
import sys
import calendar


# Complete the time_delta function below.
#Funzione che ritorna un dizionario dei mesi (chiave) con valore l'intero corrispondente 
def new_months_dict():
    l = list(calendar.month_abbr)
    d = {}
    i = 0
    for month in l:
        d[month] = i
        i += 1
    return d

#Funzione che trasforma un anno in giorni (tenendo conto dei bisestili)
def year_to_days(y):
    days = 0
    for year in range(1, y+1):
        if calendar.isleap(year):
            days += 366
        else:
            days += 365
    return days

#Funzione che trasforma un mese m di un anno y in giorni trascorsi di quello stesso anno
def month_to_days(m, y):
    days = 0
    for month in range(1, m):
        t = calendar.monthrange(y, month)
        days += t[1]
    return days

#Funzione che trasforma una data in secondi
def from_date_to_seconds(t):
    date_list = t.split()
    months_dict = new_months_dict()
    d = int(date_list[1])
    m = months_dict[date_list[2]]
    y = int(date_list[3])
    time = date_list[4].split(":")
    hours = int(time[0])
    mins = int(time[1])
    secs = int(time[2])
    zone_hrs = int(date_list[5][:3])
    zone_mins = int(date_list[5][:1]+date_list[5][3:])
    days = d + month_to_days(m, y) + year_to_days(y-1)
    return (days*(24*60*60) + (hours-zone_hrs)*(60*60) + (mins-zone_mins)*60 + secs)


def time_delta(t1, t2):
    sec1 = from_date_to_seconds(t1)
    sec2 = from_date_to_seconds(t2)
    return str(abs(sec1 - sec2))

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    t = int(input())

    for t_itr in range(t):
        t1 = input()

        t2 = input()

        delta = time_delta(t1, t2)

        fptr.write(delta + '\n')

    fptr.close()
```

```python
#XML 1 - Find the Score
def get_attr_number(node):
    # your code goes here
    sum_attrib = 0
    for child in node.iter():
        sum_attrib += len(child.attrib)
    return sum_attrib

#XML2 - Find the Maximum Depth
maxdepth = 0
def depth(elem, level):
    global maxdepth
    level += 1
    # your code goes here
    if level >= maxdepth:
        maxdepth = level
    for child in elem:
        depth(child, level)
```

```python
#Birthday Cake Candles
#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'birthdayCakeCandles' function below.
#
# The function is expected to return an INTEGER.
# The function accepts INTEGER_ARRAY candles as parameter.
#

def birthdayCakeCandles(candles):
    # Write your code here
    max_c = 0
    for elem in candles:
        if elem > max_c:
            max_c = elem
    return candles.count(max_c)

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    candles_count = int(input().strip())

    candles = list(map(int, input().rstrip().split()))

    result = birthdayCakeCandles(candles)

    fptr.write(str(result) + '\n')

    fptr.close()

#Number Line Jumps
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the kangaroo function below.
def kangaroo(x1, v1, x2, v2):
    if v1 <= v2:
        return "NO"
    while x1 < x2:
        x1 += v1
        x2 += v2
    if x1 == x2:
        return "YES"
    else:
        return "NO"

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    x1V1X2V2 = input().split()

    x1 = int(x1V1X2V2[0])

    v1 = int(x1V1X2V2[1])

    x2 = int(x1V1X2V2[2])

    v2 = int(x1V1X2V2[3])

    result = kangaroo(x1, v1, x2, v2)

    fptr.write(result + '\n')

    fptr.close()

#Viral Advertising
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the viralAdvertising function below.
def compute(n, d):
    if d == 1:
        return n//2
    return n//2 + compute(3*(n//2), d-1)

def viralAdvertising(n):
    return compute(5, n)


if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input())

    result = viralAdvertising(n)

    fptr.write(str(result) + '\n')

    fptr.close()

#Recursive Digit Sum
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the superDigit function below.
def superDigit(n, k):
    new_n = str(n)
    if(len(new_n) == 1):
        return n
    somma = 0
    for char in new_n:
        somma += int(char)
    somma = somma*k
    return superDigit(somma, 1)

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    nk = input().split()

    n = nk[0]

    k = int(nk[1])

    result = superDigit(n, k)

    fptr.write(str(result) + '\n')

    fptr.close()

#Insertion Sort - Part 1
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the insertionSort1 function below.
def insertionSort1(n, arr):
    tmp = arr[n-1]
    for i in range(n-2, -1, -1):
        if arr[i] > tmp:
            arr[i+1] = arr[i]
            arr_str = list(map(str, arr))
            print(' '.join(arr_str))
        else:
            arr[i+1] = tmp
            arr_str = list(map(str, arr))
            print(' '.join(arr_str))
            return
    arr[0] = tmp
    arr_str = list(map(str, arr))
    print(' '.join(arr_str))
    return
        

if __name__ == '__main__':
    n = int(input())

    arr = list(map(int, input().rstrip().split()))

    insertionSort1(n, arr)

#Insertion Sort - Part 2
#!/bin/python3

import math
import os
import random
import re
import sys

def insertionSort1(n, arr):
    tmp = arr[n-1]
    for i in range(n-2, -1, -1):
        if arr[i] > tmp:
            arr[i+1] = arr[i]
        else:
            arr[i+1] = tmp
            return arr
    arr[0] = tmp
    return arr

# Complete the insertionSort2 function below.
def insertionSort2(n, arr):
    for i in range(2, n+1):
        arr = insertionSort1(i, arr)
        arr_str = list(map(str, arr))
        print(' '.join(arr_str))

if __name__ == '__main__':
    n = int(input())

    arr = list(map(int, input().rstrip().split()))

    insertionSort2(n, arr)
```
